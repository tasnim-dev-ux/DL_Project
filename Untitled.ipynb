```python
import pickle
import numpy as np

# chemin vers ton fichier local
dataset_path = "data/RML2016.10a_dict.pkl"

with open(dataset_path, "rb") as f:
    data = pickle.load(f, encoding="latin1")

print(type(data))  # devrait afficher <class 'dict'>
print(len(data))   # nombre de clés/modulations
print(list(data.keys())[:5])  # voir quelques labels/modulations
```

```python
#Explorer la structure du dataset
#Combien de signaux pour chaque catégorie

#Chaque signal a normalement 2 canaux : I/Q (In-phase / Quadrature)
```

```python
# exemple : regarder un signal
example_key = list(data.keys())[0]
print("Clé :", example_key)
print("Nombre de signaux pour cette clé :", len(data[example_key]))
print("Shape d'un signal :", np.array(data[example_key][0]).shape)
```

```python
# -----------------------------------------------
# Préparation des données pour la classification
# -----------------------------------------------
# Objectif :
# Créer X et y pour entraîner un modèle CNN qui va prédire la qualité du réseau
# à partir des signaux radio I/Q.
#
# Explications :
# - X : liste des signaux I/Q
# - y : labels correspondant à la qualité réseau (Good / Average / Poor)
#
# Remarques importantes :
# - Dans ce dataset RadioML 2016.10a, les clés sont des tuples :
#   (modulation, SNR)
#   Exemple : ('QPSK', 10) → modulation QPSK avec SNR 10dB
# - On définit la qualité réseau à partir du SNR :
#     SNR >= 10   → Good
#     0 <= SNR < 10 → Average
#     SNR < 0    → Poor
# - Cette cellule prépare les données pour la prochaine étape :
#   transformation des signaux en spectrogrammes pour CNN
# -----------------------------------------------
```

```python
X = []
y = []

for key in data.keys():
    signals = data[key]
    
    modulation, snr = key  # ici key est un tuple (modulation, SNR)
    
    # Définir label qualité réseau selon SNR
    if snr >= 10:
        quality = "Good"
    elif snr >= 0:
        quality = "Average"
    else:
        quality = "Poor"
    
    for signal in signals:
        X.append(signal)
        y.append(quality)

X = np.array(X)
y = np.array(y)

print("X shape :", X.shape)
print("y shape :", y.shape)

# Vérifier la distribution des classes
from collections import Counter
print(Counter(y))
```

```python
#Vérification des valeurs aberrantes / NaN
#Comme c’est un dataset synthétique, normalement pas de NaN ni inf, 
#et valeurs centrées autour de [-1, 1] ou [-0.5, 0.5] selon la génération.
```

```python
print("Nombre de valeurs NaN dans X :", np.isnan(X).sum())
print("Nombre de valeurs infinies :", np.isinf(X).sum())
print("Valeurs min et max :", X.min(), X.max())
```
